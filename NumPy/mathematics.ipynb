```python
import numpy as np
```

```python
arr= np.array([1,2,3,4,5])
arr
```

```python
#we can do element wise operations 
arr2= arr +2
print (arr2)
arr3= arr * 2
print(arr3)
```

```python
arr4=np.array([1,1,2,3,0])
arr +arr4
```

```python
#you can also take trigo values 
np.cos(arr)
```

```python
mat1 = np.ones((2,3))
mat2= np.full((3,2), 2)
print (mat1)
print(mat2)
```

```python
#matrix multiplication 
np.matmul(mat1, mat2)
```

```python
mat3= np.identity(3)

#find determinant 
print(np.linalg.det(mat3))
```

```python
print(np.sum(arr))
#if you want a row sum or column sum then you can specify that too 
#column summed up 
print(np.sum(mat1 ,axis=0))

#row summed up 
print(np.sum(mat1, axis =1))

```
