FILTERING DATA 

```python
#athletes taller than 215 cm and born in usa 
df[(df["height_cm"] >215) & (df["born_country"] == "USA")]
```

```python
#names that start with keith 
df[df["name"].str.contains("Keith" , case = False)]
#you can do case = False whne you do not want anything to be case sensitive 
```

```python
#filtering names where it contains keith or patrick or adams
df[df["name"].str.contains("keith|patrick|adams", case = False)]
```

```python
#isin function has a list and check if any value of data frame is in the list 
# like any olympian born in use or france or ind 
df[df["born_country"].isin(['USA', "FRA", "IND"])]
```

```python
#query functions 
df.query('born_country == "IND" and born_city == "Delhi"')
```

```python
#making a copy 
bios=df.copy()
```

```python
bios["first name"]= bios["name"].str.split(' ').str[0]
```

```python
#adding birth year
bios["born_date"] = pd.to_datetime(bios["born_date"])
bios.info()
```

```python
bios["born_year"]= bios["born_date"].dt.year
```

```python
bios["age"]= 2026 - bios["born_year"]
bios.head()
```

```python
#creatinga column with our own categories using lambda functions

bios["height_category"]= bios["height_cm"].apply(lambda x: 'short' if x <165 else ('average' if x < 185 else 'tall'))
```

```python
bios.head()
```

```python
#making a custom function 
def categorize_athlete(row):
    if row["height_cm"] < 175 and row["weight_kg"] < 70 :
        return "Light Weight"
    elif row["height_cm"] < 185 and row["weight_kg"] <=80:
        return "Middle Weight"
    else:
        return "Heavy Weight"
    
#here we give axis =1 which means row 
bios["category"]= bios.apply(categorize_athlete, axis =1)
```

```python
bios.head()
```

```python
merged= pd.merge(bios,df2, left_on="born_country", right_on='NOC',how ='inner') #inner-> intersection
#here left_on-. column of left database that e are using as basis of merging 
#right_on-> column of the databse of right side 
#how -> mode of join i.e. left, right, full , inner ,outer 
```
