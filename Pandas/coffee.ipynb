```python
#to access values
#loc -> filter by rows and columns. syntax: <df_name>.loc[rows ,columns_name]
df.loc[:3 , ["Day", "Units Sold"]]
```

```python
#sorting values by columns 
df.sort_values("Units Sold", ascending = False) #sorted in descending order acc to coffee sold
```

```python
#sorting with multiple criterias 
df.sort_values(["Units Sold","Coffee Type"], ascending = [0,1]) 
#here in case of collission that two or moreunits sold rows are same it will jump to coffee type 
#to sort them out 
#0-> false (descending)
#1->true (ascending)
```

```python
#creating a copy of the dataset 
df2= df.copy()
df2.head()
```

```python
df_new= df.copy()
df_new.head()
```

```python
#adding new column
df_new["price"]= 4.99
df_new.head()
```

```python
#adding new column 
df_new["new price"]= np.where(df_new['Coffee Type']=='Espresso', 3.99, 5.99)
df_new.head()
```

```python
#renaming a column 
df_new.rename(columns={"new price":"updated price"}, inplace = True)
df_new.head()
```

```python
#dropping a column 
df_new.drop(columns=['price'], inplace = True)
```

```python
#maing a new column from existing column 
df_new["revenue"]= df_new["Units Sold"] * df_new["updated price"]
df_new.head()
```

HANDLING NULL VALUES 

```python
#setting some values null 
df.loc[[2,3], 'Units Sold'] = np.nan
```

```python
#getting info on null values
df.isna().sum()
```

```python
#filling null values 
#df.fillna(df['Units Sold'].mean())

#we can also use interpolate instead of mean 
#interpolate -> checks if there is a pattern in the data and fill null values accordingly
df['Units Sold'].interpolate()
```

```python
#we can directly drop the rows with null values using 
df.dropna()

#if you want to drop only when some specific column value is null then you can do 
df.dropna(subset=['Units Sold'], inplace = True)

#we can get rows with null values 
df[df['Units Sold'].isna()]

#the opposite that is getting rows which are not null then 
df[df['Units Sold'].notna()]
```

GROUPBY FUNCTION

```python
df_new.head()
```

```python
#finding how much units were sold by each type of coffee
df_new.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
#we can also perform different functions on different columns 
df_new.groupby(['Coffee Type']).agg({'Units Sold': 'sum', 'updated price': 'mean'})
```

```python
df_new = df_new.fillna(df_new['Units Sold'].mean())
df_new.head()
```

```python
#we can also reform the table according to our own needs 
#that is done by pivot 
pivot = df_new.pivot(columns='Coffee Type', index ='Day', values ='revenue')
```

```python
#a dataframe was made where columns were created on the basis of different values of the main column we passed 
#initialization. and the index and values are as we set giving us a custom data frame
pivot
```

```python
#giving sum of all the columns
pivot.sum()
```

```python
#giving sum of all the rows
pivot.sum(axis =1)
```

ADVANCED FUNCTIONS 


```python
#.shift() -> used to get previous values 
df_new ['yesterday_revenue']= df_new['revenue'].shift(1)
df_new.head()
```

```python
#rolling fucntions like cumulative sum 
df_new['cumulative_revenue']= df_new['revenue'].cumsum()
df_new.head()
```
