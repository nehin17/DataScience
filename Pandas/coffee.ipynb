```python
import pandas as pd
```

```python
#reading a csv file
df=pd.read_csv('coffee.csv')
df
```

```python
df.head()
```

```python
#parquet file are used as they take less space so they improve speed and performance the downside 
#is that they are not very readable like csv or excels
```

```python
#to get some random samples you can do 
df.sample(4)
```

```python
#to access values
#loc -> filter by rows and columns. syntax: <df_name>.loc[rows ,columns_name]
df.loc[:3 , ["Day", "Units Sold"]]
```

```python
#iloc -> uses index instead of column names
#syntax: <df_name>.loc[rows ,columns_index]
df.iloc[5:9,[0,2]]

'''one difference between loc and iloc is that in loc the upper index of the row is inclusive 
but in iloc he uppper index ofrow is exclusive'''
```

```python
#we can change the index (serial number) of the dataframe and put it the same as any of the column

#for example putting day column as index 
df.index= df["Day"]
df.head()
```

```python
#modifying values
df[1, "Units Sold"]= 20
df.head()
```

```python
#getting specifc values 
df.at[3,"Units Sold"]
```

```python
#getting a specific column
df["Coffee Type"]
```

```python
#sorting values by columns 
df.sort_values("Units Sold", ascending = False) #sorted in descending order acc to coffee sold
```

```python
#sorting with multiple criterias 
df.sort_values(["Units Sold","Coffee Type"], ascending = [0,1]) 
#here in case of collission that two or moreunits sold rows are same it will jump to coffee type 
#to sort them out 
#0-> false (descending)
#1->true (ascending)
```

```python
#iterating the database using for loop 
for index, row in df.iterrows():
    print(index)
    print(row["Day"])
    print("\n")
```
